## 비교 연산자

```python
a = 3
if a:
    print('hahaha')
```

* 0(혹은 None)이 아닌 숫자 값은 다 true

```python
a = 3.13
a += 0.01

e = 1e-6

abs(a - 3.14) < e
```

* 실수는 비교 어려움

```python
s1 = 'hello'
s2 = 'hello'

s1 == s2
```

* 문자열 equals() 메서드 쓰지 않아도 비교 가능

```python
d1 = [1, 2, 3]
d2 = [1, 2, 3]

d1 == d2
```

```python
id(d1) == id(d2)
```

* 배열은 각 요소를 비교해서 판별함
* id 값은 같지 않음

```python
'h' in s1, 3 in d1, 4 in d1
```

```python
a = 1
b = 2

if (a == 1) and b == 2:
    pass
```

```python
a = 1
b = 2

if not (a == 1 and b == 2):
    pass
```

## if문

* 실행문과 조건문을 한 줄에 적는 것도 가능

```python
b = 'ok' if a > 5 else 'no'
b
```

```python
data = [1, 2, 3, 4, 5]
[d for d in data if d % 2 == 0]
```

```python
a = 1
b = 3

if a == 1:
    if b == 2:
        print('a')
    print('b') # 들여쓰기에 따라 출력되는 값이 달라짐
```

## 정렬

```python
data = [2, 1, 3]
data.sort()
data # 원본이 변경됨, 반환값은 None
```

```python
data = [2, 1, 3]
sorted(data), data, sorted(data, reverse=True) # 원본은 변경되지 않음, 반환값은 정렬된 새로운 리스트
```

```python
max(3, 1, 2) # 인수 여러 개 줄 수 있음
```
