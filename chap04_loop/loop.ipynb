# 반복문   

## while

```python
a = 5
while a > 0:
    print(a)
    a -= 1
```

```python
# 1부터 5까지의 합
start = 1
end = 5
result = 0
while start <= end:
    result += start
    start += 1
result
```

```python
data = [1, 2, 3, 4, 5]
result = 0
for a in data:
    result += a
result
```

```python
# 1부터 5까지 홀수끼리 더하기
start = 1
end = 5
result = 0
while start <= end:
    if (start % 2 == 0):
        start += 1
        continue
    result += start
    start += 1
result
```

```python
# 5까지 홀수만 출력
i = 0
start = 1
end = 5
while i < 100:
    if start > end: break
    if start % 2 == 0:
        start += 1
        continue
    print(start)
    start += 1
    i += 1
```

## 난수

```python
import random

# 실수 난수
for _ in range(10):
    print(random.random())
```

```python
# 정수 난수
random.randint(1, 6)
```

```python
data = []
for _ in range(10):
    data.append(random.random())
    print(data[_])
```

## for

```python
data = list(range(11))
data
```

```python
for d in data:
    print (d)
```

```python
# range는 클래스, 함수처럼 생겼지만 생성자 호출 중
for d in range(1, 11): # range(start, end, step)
    print (d)
```
