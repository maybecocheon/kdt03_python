# 반복문   

## while

```python
a = 5
while a > 0:
    print(a)
    a -= 1
```

```python
# 1부터 5까지의 합
start = 1
end = 5
result = 0
while start <= end:
    result += start
    start += 1
result
```

```python
data = [1, 2, 3, 4, 5]
result = 0
for a in data:
    result += a
result
```

```python
# 1부터 5까지 홀수끼리 더하기
start = 1
end = 5
result = 0
while start <= end:
    if (start % 2 == 0):
        start += 1
        continue
    result += start
    start += 1
result
```

## 난수

```python
import random

# 실수 난수
for _ in range(10):
    print(random.random())
```

```python
# 정수 난수
random.randint(1, 6)
```

```python
data = []
for _ in range(10):
    data.append(random.random())
    print(data[_])
```
