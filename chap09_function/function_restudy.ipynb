```python
def add (a: int, b: int) -> int:
    """
    
    더하기 함수입니다.

    인자:
    a: 정수
    b: 정수

    반환값:
    덧셈의 결과로 정수가 반환됨

    """
    if isinstance(a, int) and isinstance(b, int):
        return a + b
    raise Exception('입력된 값이 정수가 아닙니다.')
```

```python
# p. 430 윤년 구하기

"""특정 해의 일수 구하기"""
def is_leapyear(year) -> bool:
    return True if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0 else False

print('특정 해의 일수를 구합니다.')
a = input('특정 해를 입력하세요.')
print(f'{a}년의 일수는 {366 if is_leapyear(int(a)) else 365}일입니다.')
```

```python
def clean(s):
    return s.replace(" ", "")

def clean2(s):
    import re
    import string
    return re.sub(f'[{string.whitespace}]', "", s)
```
