```python
# p. 385 직각 이등변 삼각형과 사각형

def put_star(n):
    """n개의 '*'을 연속해서 출력"""     # docstring: help() 메서드 사용하면 읽을 수 있는 주석
    for _ in range(n):
        print('*', end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    put_star(i)
    print()

print('사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    put_star(w)
    print()
```

```python
# p.390 두 값의 최솟값과 최댓값 구하기

def min_max2(a, b):
    """a와 b의 최솟값과 최댓값을 구하여 리턴"""
    return (a, b) if a < b else (b, a)

n1 = int(input('정수 n1：'))
n2 = int(input('정수 n2：'))

minimum, maximum = min_max2(n1, n2)     # 언팩
print('최솟값은', minimum, '이고 최댓값은', maximum, '입니다.')
```

```python
# p.397 인수 교환 메커니즘
# 1부터 n까지의 합 구하기

def sum_1ton(n):
    """1부터 n까지의 합 구하기"""
    s = 0
    while n > 0:
        s += n
        n -= 1
    return s

x = 5
total = sum_1ton(x)
print('1부터 ', x, '까지의 합은 ', total, '입니다.', sep='')
```

```python
# p. 400 리스트를 인수로 받는 함수
# 리스트의 요소 값 변경하기 => 리스트는 가변 객체라서 요소 변경하고도 동일 id값 가짐

def change(lst, idx, val):
    """lst[idx]의 값을 val으로 변경"""
    lst[idx] = val

x = [11, 22, 33, 44, 55]
print('x =', x, id(x))

index = 2
value = 99

change(x, index, value)
print('x =', x, id(x))
```

```python
# 리스트를 슬라이싱으로 전달

def change(lst, idx, val):
    """lst[idx]의 값을 val으로 변경"""
    lst[idx] = val

x = [11, 22, 33, 44, 55]
print('x =', x, id(x))

index = 2
value = 99

y = x[:]
change(y, index, value)
print('x =', x, id(x))      # 슬라이싱은 사본을 만들어 전달하기 때문에 원본 변경 x => 함수가 리스트를 전달받아 원본 수정 못 하게 하려면 슬라이싱 사용하여 전달
print('y =', y, id(y))
```

```python
# 직각 이등변 삼격형과 사각형 출력하기1(위치 인수)

def puts(n, s):
    """n개의 s를 연속해서 출력"""
    for _ in range(n):
        print(s, end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    puts(i, '*')
    print()

print('사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    puts(w, '+')        
    print()
```

```python
# 직각 이등변 삼격형과 사각형 출력하기2(키워드 인수)

# n개의 s를 연속해서 출력
def puts(n, s):
    for i in range(n):
        print(s, end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    puts(n = i, s = '*')
    print()

print('사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    puts(s = '+', n = w)        # 키워드 인수: 함수 호출 시 매개변수 = 값
    print()
```

```python
# 두 개 이상의 값 중 최댓값 구하기

def max2more(a, b, *num):           # * 하면 num을 튜플로 받음
    """두 개 이상의 값 중 최댓값 구하기"""
    max = a if a > b else b
    for n in num:
        if n > max:
            max = n
    return max

print('max2more(1, 2)          = ', max2more(1, 2))
print('max2more(1, 2, 3)       = ', max2more(1, 2, 3))
print('max2more(1, 2, 3, 4, 5) = ', max2more(1, 2, 3, 4, 5))
print('max2more(1)             = ', max2more(1))
```
