```python
# p. 385 직각 이등변 삼각형과 사각형

def put_star(n):
    """n개의 '*'을 연속해서 출력"""     # docstring: help() 메서드 사용하면 읽을 수 있는 주석
    for _ in range(n):
        print('*', end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    put_star(i)
    print()

print('사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    put_star(w)
    print()
```

```python
# p.390 두 값의 최솟값과 최댓값 구하기

def min_max2(a, b):
    """a와 b의 최솟값과 최댓값을 구하여 리턴"""
    return (a, b) if a < b else (b, a)

n1 = int(input('정수 n1：'))
n2 = int(input('정수 n2：'))

minimum, maximum = min_max2(n1, n2)     # 언팩
print('최솟값은', minimum, '이고 최댓값은', maximum, '입니다.')
```

```python
# p.397 인수 교환 메커니즘
# 1부터 n까지의 합 구하기

def sum_1ton(n):
    """1부터 n까지의 합 구하기"""
    s = 0
    while n > 0:
        s += n
        n -= 1
    return s

x = 5
total = sum_1ton(x)
print('1부터 ', x, '까지의 합은 ', total, '입니다.', sep='')
```

```python
# p. 409 위치 인수를 튜플로 만들어 가변 인수 전달하기
# 두 개 이상의 값 중 최댓값 구하기

def max2more(a, b, *num):           # * 하면 num을 튜플로 받음
    """두 개 이상의 값 중 최댓값 구하기"""
    max = a if a > b else b
    for n in num:
        if n > max:
            max = n
    return max

print('max2more(1, 2)          = ', max2more(1, 2))
print('max2more(1, 2, 3)       = ', max2more(1, 2, 3))
print('max2more(1, 2, 3, 4, 5) = ', max2more(1, 2, 3, 4, 5))
print('max2more(1)             = ', max2more(1))
```

```python
# p. 412 이터러블형 인수의 언팩
# 리스트에 있는 세 값의 최댓값을 구하기

def max3(a, b, c):
    """a, b, c의 최댓값을 구해 리턴"""
    max = a
    if b > max: max = b
    if c > max: max = c
    return max

lst1 = [1, 3, 5]
m = max3(*lst1)
print(lst1, '의 최댓값은', m, '입니다.')
```

```python
# p. 414
# 인수의 합을 출력하면서 구하기(인수 언팩)

def print_sum(a, *no):      # 팩 - 튜플로 나머지 인수를 받는다
    """인수의 합을 리턴(식도 출력)"""
    sum = a
    print(a, end='')
    n = len(no)
    if n > 0:
        print(' + ', end='')
        for i in range(n - 1):
            sum += no[i]
            print(no[i], '+ ', end='')
        sum += no[n - 1]
        print(no[n - 1], end='')
    print(' =', sum)
    return sum

lst1 = [1, 3, 5, 7]
print_sum(*lst1)        # 언팩 - 리스트의 값을 a, *no에 나눠서 보냄
```

```python
# p. 415 키워드 인수를 딕셔너리로 전달하기
# 딕셔너리화된 키워드 인수 정보 출력하기

def print_kwargs(s, **kwargs):          # 팩 - 딕셔너리로 팩킹
    """딕셔너리화된 키워드 인수를 받는 kwargs의 정보 출력하기"""
    print(s)
    print('type(kwargs) =', type(kwargs))
    print('len(kwargs)  =', len(kwargs))
    print('kwqrgs       =', kwargs)

print_kwargs('1번', spring='봄', summer='여름')
print()
print_kwargs('2번', spring='봄')
```

```python
# p. 416
# 딕셔너리에 저장된 사람의 정보 출력하기

def put_person(**person):           # ** - 키워드 인수를 팩 하는 것
    """딕셔너리 person에 있는 정보를 출력"""
    if 'name' in person: print('이름 =', person['name'], end='  ')
    if 'visa' in person: print('국적 =', person['visa'], end='  ')
    if 'age'  in person: print('연령 =', person['age'],  end='  ')
    print()  # 개행

put_person(name='손지형', visa='한국', age=27)
put_person(name='조탁', visa='중국')
put_person(**{'name':'손지형', 'visa':'한국', 'age':27})
```

```python
# p. 418
# 딕셔너리형 인수를 **로 언팩하기

def puts(n, s):         # 딕셔너리의 키로 매핑 해서 씀
    """n개의 s를 연속해서 출력"""
    for _ in range(n):
        print(s, end='')

d1 = {'n': 3, 's': '*'}     # '*' 3개
d2 = {'s': '+', 'n' :7}     # '+' 7개

puts(**d1)      # 딕셔너리 언팩 => puts('n': 3, 's': '*') 형식이 아니라 puts(n = 3, s = '*') 형식(키워드 인수)으로 전달하는 것
print()
puts(**d2)
```

```python
# p. 426
# 어노테이션이 있는 함수 puts

def puts(n: int, s: str) -> None:       # 1) : 타입 => 어노테이션   2) -> 타입 => 어노테이션
    """n개의 s를 연속해서 출력"""
    for _ in range(n):
        print(s, end='')

puts(5, '*')
print()
print(puts.__annotations__)     # 어노테이션을 알 수 있음
print()
puts('*', 5)
```

```python
# p. 429 docstring의 표준 코딩 스타일


"""어노테이션과 문서화 문자열이 있는 함수 puts"""


def puts(n: int, s: str) -> None:
    """n개의 s를 연속해서 출력

    매개 변수:
        n -- 출력하는 문자열의 개수
        s -- 출력하는 문자열
    리턴값:
        없음

    """
    for _ in range(n):
        print(s, end='')


print(puts.__doc__)    # 문서화 문자열을 출력
```

```python
# p. 431
"""딕셔너리에서 특정한 값을 가지는 키의 리스트 생성"""

def keys_of(dic: dict, val: 'value') -> list:
    """딕셔너리 dic에 있는 값이 val인 요소의 키를 리스트로 리턴"""
    return [k for k, v in dic.items() if v == val]

txt = input('문자열 : ')
count = {ch: txt.count(ch) for ch in txt}
print('분포=', count)

num = int(input('몇 개짜리 문자：'))
print('{}개의 문자={}'.format(num, keys_of(count, num)))
```

```python
# p. 434
"""곱셈 표 출력"""

upper = int(input('1부터 몇까지 곱할까요：'))

def multiplication_table(n: int) -> bool:
    """1 ~ n까지의 곱셈 표를 출력"""
    if    1 <= n <=  3: w = 2
    elif  4 <= n <=  9: w = 3
    elif 10 <= n <= 31: w = 4
    else              : return False

    f = '{{:{}d}}'.format(w)            # {{:3d}} => 중괄호 중첩 시 안에 있는 문자 그대로 '{:3d}' 쓰겠다는 뜻 => 세 칸마다 띄어 쓰겠다는 뜻
    print('-' * n * w)
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            print(f.format(i * j), end='')
        print()
    print('-' * n * w)
    return True

multiplication_table(upper)
```

```python
# p. 437 내부 함수
"""곰셈 표 출력(내부 함수)"""

upper = int(input('1부터 몇까지 곱할까요：'))

def multiplication_table(n: int) -> bool:
    """1 ~ n까지의 곱셈 표를 출력"""

    def put_bar(n: int) -> None:
        """n개의 '-'을 연속해서 출력하고 개행"""
        print('-' * n)
        print('내부함수 n:', id(n))

    if    1 <= n <=  3: w = 2
    elif  4 <= n <=  9: w = 3
    elif 10 <= n <= 31: w = 4
    else              : return False

    f = '{{:{}d}}'.format(w)
    put_bar(n * w)
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            print(f.format(i * j), end='')
        print()
    put_bar(n * w)
    print('외부함수 n:', id(n))
    return True

multiplication_table(upper)
```

```python
# p. 439 글로벌 네임스페이스와 로컬 네임스페이스
# 변수의 스코프(영역) 확인
"""변수 내외의 변수 값을 확인"""

g1 = 1
g2 = 2

def outer():
    def inner():
        f2 = 6
        print('[2] inner() g1, g2 = ', g1, g2)
        print('    inner() f1, f2 = ', f1, f2)

    f1 = 3
    f2 = 4
    g2 = 5      # 글로벌 변수와 같은 이름을 가지지만 로컬변수로써 다른 변수를 생성한 것임
    inner()
    print('[3] outer() g1, g2 = ', g1, g2)
    print('    outer() f1, f2 = ', f1, f2)

print('[1] global  g1, g2 = ', g1, g2)
outer()
print('[4] global  g1, g2 = ', g1, g2)
```

```python
# nonlocal문 있음

def outer():
    n = 1
    def inner():
        nonlocal n
        n = 2
        print('n =', n)

    print('n =', n)
    inner()
    print('n =', n)

outer()
```

```python
# p. 447
"""함수가 객체인지 확인"""

def min2(a, b):
    """a, b의 최솟값을 구해 리턴하는 함수"""
    return a if a < b else b

a = int(input('정수 a：'))
b = int(input('정수 b：'))

func = min2                                   # 함수는 객체 > 참조변수가 가리킬 수 있음
print('최솟값은', func(a, b), '입니다.')        # 참조변수가 함수 객체를 가리키면 참조변수로 함수 호출 가능

del min2
print('최솟값은', func(a, b), '입니다.')

del func
print('최솟값은', func(a, b), '입니다.')
```

```python
# p. 451
"""9행 9열의 곱셈과 덧셈표 출력하기"""

def gugu(func):     # 고차원 함수
    """9행 9열의 표를 출력"""
    for i in range(1, 10):
        for j in range(1, 10):
            print('{:3d}'.format(func(i, j)), end='')
        print()

def mul2(x, y):
    return x * y

def add2(x, y):
    return x + y

n = int(input('곱셈[0]/덧셈[1]：'))

if n == 0:
    print('9행 9열의 곱셈 표')
    gugu(mul2)
elif n == 1:
    print('9행 9열의 덧셈 표')
    gugu(add2)
```

```python
# p. 458 filter 함수와 람다식
# 80점 이상의 점수만 추출하기

import random

number = int(input('학생 수 : '))

score = [None] * number

for i in range(number):
    score[i] = random.randint(0, 100)

print('모든 학생 점수＝', score)
print('합격자 점수', list(filter(lambda n: n >= 80, score)))
```
