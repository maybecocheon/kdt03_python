```python
# p. 259
x = [15, 64, 7, 3.14, [32, 55], 'ABC']
print('리스트: ', x)
print('리스트의 길이: ', len(x))
print('리스트의 타입: ', type(x))
print('인덱스로 접근: ', x[4])
print('리스트 내 리스트를 인덱스로 접근: ', x[4][0]) # 요소도 객체
```

```python
# p.261 리스트 생성
# 방법 1: [] 사용
list2 = [1, 'good']
print('list2: ', list2)

# 방법 2: list() 함수 사용
list3 = list(range(6))
print('list3: ', list3)
list5 = list(range(6))
print('list5: ', list5)

# 방법 3: 요소 개수 지정하여 생성
list4 = [None] * 5           # 5개 요소를 생성
print('list4: ', list4)

# 리스트의 동일성 비교
print('id값 비교: ', list3 is list5)        # id값(주소값) 다르기 때문에 False 나옴
print('요소 비교: ', list3 == list5)        # 요소 같기 때문에 True 나옴
```

```python
# p. 263 리스트 대입
lst1 = list3         # 치환문 => list3 참조변수 값을 lst1도 참조하도록 한 것
print('lst1: ', lst1)
lst1[2] = 99         # 치환했기 때문에 원본 변경됨
print('lst3: ', list3)
print('id값 비교: ', lst1 is list3)
```

```python
l1 = [1, 2, 3]
l2 = [4, 5]
x = l1 + l2
print('l1 + l2 = x', x)
x[2] = 99           # x는 메모리를 새로 만들어서 참조하는 것이기 때문에 원본 변경되지 않음
print('요소 변경 후 x: ', x)
print('l1 값 변경 없음: ', l1)
print('l1 값 변경 없음: ', l2)
x2 = x * 2
print('x * 2 = ', x2)
```

```python
# p. 267 리스트 언팩
x = [1, 2, 3]
a, b, c = x     # 리스트의 각 요소를 풀어서 변수에 할당
print('a의 값: ', a)

# 다중 치환
d, e, f = 10, 11, 12
print('d의 값: ', d)
```

```python
# p.268 인덱스 사용과 슬라이싱
x = [11, 22, 33, 44, 55, 66, 77]
print('끝에서부터 두 번째: ', x[-2])
print('끝에서부터 세 번째에서 맨 뒤까지: ', x[-3:])
print('처음부터 맨끝에서 두 번째까지', x[:-1])
print('역순: ', x[::-1])
```

```python
# p. 271 리스트 검색
print('x에 11의 포함 여부: ', 11 in x)      # 귀속 판정 연산자
a = 1
b = [1, 2]
c = [[1, 2], [3, 4]]
print('a는 b 안에, b는 c 안에 포함 여부: ', a in b in c)    # (a in b) and (b in c)로 해석
x = [n for n in range(6)]       # 리스트 내포 표기 p. 302
print(x)
```

```python
# p. 274 리스트 확장
x = [11, 22, 33, 44, 55, 66, 77]
x.append(88)
print('append 메서드: ', x)
x3 = [11, 22, 33, 44, 55, 66, 77]
x3.extend([88, 99])
print('extend 메서드: ', x3)
x2 = x * 2
print('x * 2 = ', x2)
y = [11, 22, 33]
y *= 3
print('y 누적 대입(제자리 연산): ', y)
```

```python
# p.276
x = x + y
x += y      # 누적 대입 시 제자리 연산 => 원본 변경
```

```python
# p. 277 요소 삽입과 삭제
x = list(range(11, 78, 11))
print('x 리스트: ', x)
# remove
y = x.remove(77)
print('값 77 삭제(remove): ', y)
# pop
v = x.pop(2)
print('인덱스 2 값 삭제(pop): ', v)
# del문
del x[2] # 인덱싱 사용한 삭제
print('remove, pop, del문 사용 후 x 리스트: ', x)
```

```python
# p. 281 리스트 이터러블
# 1.  리스트의 모든 요소 탐색(요소 개수 미리 확인)
x = ['john', 'geoge', 'judy', 'tom']
print('[len으로 리스트 요소 탐색]')
for i in range(len(x)):
    print(x[i])

print()

# 2. enumerate 함수로 탐색
print('[enumerate 함수로 리스트 요소 탐색]')
for idx, name in enumerate(x):          # 정말 많이 사용함. 인덱스와 값을 반환.
    print(f'x[{idx}] = {name}')

print()

# 2-1. enumgerate 시작할 인덱스 값 주기
print('[enumerate 함수 시작할 인덱스 값 5]')
for idx, name in enumerate(x, 5):         
    print(f'x[{idx}] = {name}')

print()

# 3. 인덱스 사용 없이 요소 탐색
print('[인덱스 없이 리스트 요소 탐색]')
for i in x:
    print(i)
```

```python
import copy
# p. 298 리스트 복사
x = [[1, 2, 3], [4, 5, 6]]
print('x 행렬:', x)
x2 = x          # 복사가 아님. 참조변수 x2가 x가 가리키는 값을 가리키는 것
x2[0][0] = 9
print('x2[0][0] 값 변경 후 x 행렬:', x)

print('\n얕은 복사')
x3 = x.copy()
x3[0][0] = -1
print('x3[0][0] 값 변경 후 x 행렬:', x)

x7 = list(x)
x7[0][0] = 99
print('x7[0][0] 값 변경 후 x 행렬:', x)

x8 = x[:]       # 함수에서 많이 사용하므로 기억
x8[0][0] = 999
print('x8[0][0] 값 변경 후 x 행렬:', x)

# 깊은 복사
print('\n깊은 복사')
x4 = copy.deepcopy(x)
x4[0][0] = 1
print('x4[0][0] 값 변경 후 x4 행렬:', x4)
print('x4[0][0] 값 변경 후 x 행렬:', x)

# 3차원 행렬
print('\n3차원 행렬 깊은 복사')
x5 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
x6 = copy.deepcopy(x5)
x6[0][0][0] = 99
print('x6[0][0][0] 값 변경 후 x6 행렬:', x6)
print('x6[0][0][0] 값 변경 후 x5 행렬:', x5)
```

```python
# p. 301 리스트 내포 표기
x = [n for n in range(1, 8)]
print(x)

# 식 부분 수정
x2 = [f'{n}{'+' if n % 2 == 1 else '-'}' for n in range(1, 8)]
print('숫자 뒤에 짝수면 -, 홀수면 + 붙이기:', x2)

# 조건문 첨가
x3 = [n for n in range(1, 8) if n % 2 == 0]
print('짝수:', x3)

# 중첩 내포 표기
x4 = [i * 10 + j for i in range(1, 3) for j in range(1, 4)]         # 앞선 반복문을 먼저 돈다
print(x4)

x5 = [[j * 10 + i for j in range(1, 3)] for i in range(1, 4)]       # 외부 반복문을 먼저 돈다
print(x5)

x6 = [[1 if j == i else 0 for j in range(1, 3)] for i in range(1, 4)]
print(x6)
```
