## 리스트 검색

```python
# 값으로 인덱스 검색
data = [1, 2, 3, 2]
data.index(2), data.index(2, 2) # 값(2)이 두 번째로 있는 인덱스
```

## 리스트 삽입

```python
data = [1, 2, 3]
data.extend([3, 1]) # data + [3, 1] => 리스트를 요소로 나눠서 추가
data
```

```python
data = [1, 2, 3]
data.append([3, 1]) # 요소 그 자체로 추가
data
```

```python
x = [1, 2, 3]
id(x)
```

```python
x = x + [4, 5]
id(x) # id 값 변경됨
```

```python
x += [4, 5]
id(x) # id 값 변경 안 됨 => 제자리 연산
```

## 리스트 역순

```python
x = [1, 2, 3]
for i in sorted(x, reverse=True):
    print(i)
```

```python
x = [1, 2, 3]
x[::-1]
```

```python
x = [1, 2, 3]
x.reverse() # 원본 수정 => 제자리 연산(in-place)
x
```

```python
x = [1, 2, 3]
y = list(reversed(x)) # 원본 수정 x
x, y
```
