## 실습 문제: 섭씨-화씨 양방향 온도 변환기 제작 및 테스트

**설명**: 이전 실습을 확장하여, 이번에는 섭씨를 화씨로, 그리고 화씨를 섭씨로 변환하는 두 개의 개별 함수를 만듭니다. 함수를 기능 단위로 분리하는 연습을 하고, 각 함수가 올바르게 작동하는지 확인하기 위한 간단한 테스트 코드를 작성하여 모듈화된 프로그래밍의 기초를 학습합니다.

$$F = C \times \frac{9}{5} + 32 \quad \text{그리고} \quad C = (F - 32) \times \frac{5}{9}$$

**요구사항**:

  - **함수 1**: `celsius_to_fahrenheit` 함수를 완성하여 섭씨 온도를 화씨 온도로 변환하세요.
  - **함수 2**: `fahrenheit_to_celsius` 함수를 완성하여 화씨 온도를 섭씨 온도로 변환하세요.
  - **테스트 코드**: 제공된 테스트 코드 섹션에서 각 함수를 호출하여 변환이 정확한지 확인하고 결과를 출력하세요. (예: 0°C → 32°F, 212°F → 100°C)



```python

def celsius_to_fahrenheit(celsius: float) -> float:
    """섭씨(Celsius)를 화씨(Fahrenheit)로 변환합니다."""
    # 이곳에 섭씨 -> 화씨 변환 코드를 작성하세요.
    return celsius * (9. / 5) + 32

def fahrenheit_to_celsius(fahrenheit: float) -> float:
    """화씨(Fahrenheit)를 섭씨(Celsius)로 변환합니다."""
    # 이곳에 화씨 -> 섭씨 변환 코드를 작성하세요.
    return (fahrenheit - 32) * (5. / 9)
```

```python
# 테스트 1: 0°C가 32°F로 변환되는지 확인
c_temp = 0
f_expected = 32
f_actual = celsius_to_fahrenheit(c_temp)
print(f"{c_temp}°C -> {f_actual}°F (예상: {f_expected}°F)")
print(f"테스트 통과: {f_actual == f_expected}")

print("-" * 20)

# 테스트 2: 212°F가 100°C로 변환되는지 확인
f_temp = 212
c_expected = 100
c_actual = fahrenheit_to_celsius(f_temp)
print(f"{f_temp}°F -> {c_actual}°C (예상: {c_expected}°C)")
print(f"테스트 통과: {c_actual == c_expected}")
```
