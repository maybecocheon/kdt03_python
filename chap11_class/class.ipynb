```python
def deco(func):
    def wrapper(*args, **kwargs):
        print(f"[LOG] í•¨ìˆ˜ {func.__name__} í˜¸ì¶œ")
        result = func(*args, **kwargs)
        print(f"[LOG] í•¨ìˆ˜ {func.__name__} ì¢…ë£Œ")
        return result
    return wrapper
```

```python
class Member:
    def __init__(self, name, weight):
        self._name = name
        self._weight = weight

    # -------- getter í•¨ìˆ˜ --------
    def get_weight(self):
        print("âš™ getter í˜¸ì¶œ")
        return self._weight

    # -------- setter í•¨ìˆ˜ --------
    def set_weight(self, w):
        print("âš™ setter í˜¸ì¶œ")
        if w > 0:
            self._weight = w
        else:
            self._weight = 0

    # -------- deleter í•¨ìˆ˜ --------
    def del_weight(self):
        print("âš™ deleter í˜¸ì¶œ")
        del self._weight

    # weight í”„ë¡œí¼í‹° ìƒì„±
    weight = property(get_weight, set_weight, del_weight, doc="ì²´ì¤‘ í”„ë¡œí¼í‹°")
```

```python
kim = Member("ê¹€ì¤€í˜¸", 72.7)

print(kim.weight)      # getter í˜¸ì¶œ
kim.weight = 67.7      # setter í˜¸ì¶œ
del kim.weight         # deleter í˜¸ì¶œ
```

```python
class Member:
    def __init__(self, name, weight):
        self._name = name
        self._weight = weight

    @property
    def weight(self):
        print("âš™ getter í˜¸ì¶œ")
        return self._weight

    @weight.setter
    def weight(self, w):
        print("âš™ setter í˜¸ì¶œ")
        self._weight = w if w > 0 else 0

    @weight.deleter
    def weight(self):
        print("âš™ deleter í˜¸ì¶œ")
        del self._weight
```

```python
class Member:
    def __init__(self, no, name, weight):
        self.no = no
        self.name = name
        self.weight = weight

    def __repr__(self):
        # ì´ ë¬¸ìì—´ì´ Member ê°ì²´ ìƒì„±ì‹ ê·¸ëŒ€ë¡œì—¬ì•¼ í•œë‹¤
        return f"Member({self.no!r}, {self.name!r}, {self.weight!r})"
```

```python
class Member:
    def __init__(self, no, name, tags):
        self.no = no
        self.name = name
        self.tags = tags

    def __repr__(self):
        return f"Member({self.no!r}, {self.name!r}, {self.tags!r})"

m1 = Member(10, "Alice", ["í—¬ìŠ¤", "ìš”ê°€"])
copy = eval(repr(m1))

print(m1)
print(copy)
print(m1.tags == copy.tags)   # True
print(m1 is copy)             # False ë³„ê°œ ê°ì²´
```

```python
class Member:
    total_count = 0   # í´ë˜ìŠ¤ ë³€ìˆ˜ (ëª¨ë“  ê°ì²´ê°€ ê³µìœ )

    def __init__(self, name):
        self.name = name
        Member.total_count += 1

    # ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ â†’ selfë¡œ ê°œë³„ ê°ì²´ì— ì ‘ê·¼
    def hello(self):
        print(f"ì•ˆë…•í•˜ì„¸ìš”, ì €ëŠ” {self.name}ì…ë‹ˆë‹¤.")

    # í´ë˜ìŠ¤ ë©”ì†Œë“œ â†’ clsë¡œ í´ë˜ìŠ¤ ìì²´ì— ì ‘ê·¼
    @classmethod
    def show_total(cls):
        print(f"íšŒì› ìˆ˜: {cls.total_count}ëª…")

    # ì •ì  ë©”ì†Œë“œ â†’ í´ë˜ìŠ¤/ê°ì²´ì™€ ë¬´ê´€, ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
    @staticmethod
    def is_valid_name(name):
        return len(name) >= 2
```
