```python
# 오버로딩 안 됨 => 덮어 씀
class Num:
    # def __init__(self):
    #     self.n = 0

    def __init__(self, n = 0):
        self.n = n

    def __str__(self):
        return str(self.n)
    
    def __add__(self, other):
        return Num(self.n + other.n)        # => 객체로 리턴할 수 있게
    
    def __mul__(self, other):
        pass
```

```python
num1 = Num()
print(num1)
num2 = Num(2)
print(num1 + num2)
```

```python
# 상속
class EvenNumber(Num):
    def __init__(self, n = 0, m = 0):
        super().__init__(n)
        self.m = m

    def __str__(self):
        return f'{self.n}:{self.m}'
    
    def __add__(self, other):
        return EvenNumber(self.n + other.n)
    
    def __sub__(self, other):
        return EvenNumber(self.n - other.n)
```

```python
print(EvenNumber(4) + EvenNumber(2))
print(EvenNumber(4) - EvenNumber(2))
print(Num())
```

```python
class my_range:
    def __init__(self, start, end):
        self.n = end
        self.cur = start

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.cur >= self.n:
            raise StopIteration()
        cur = self.cur
        self.cur += 1
        return cur
    
    def __call__(self):
        return self.__next__()
```

```python
print(my_range(1, 10))
ran = my_range(1, 10)
iter(ran)
next(ran)
next(ran)
next(ran)
next(ran)
```

```python
for i in my_range(1, 10):
    print(i)
```
