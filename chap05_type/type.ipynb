# 객체와 타입

## 객체

```python
a = 1
a = 3.14
a = 'hello'
a = True
# a가 각각 다른 메모리를 가리키는 것
```

```python
s = 'hello'
s = s + 'world'
print(s)
# s는 새로운 메모리를 만들어서 가리키는 것 (str, 숫자형은 불변형)
```

```python
data = [1, 2, 3]
del data[1]
data
# del => 배열 요소의 메모리 주소 지우는 것
```

```python
data = [1, 2, 3]
data.remove(1)
data

# 리스트에서 특정 값을 가진 첫 번째 요소를 삭제
```
