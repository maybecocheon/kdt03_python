## enumerate 함수

```python
a = 'hello'
b = [1, 2, 3, 4, 5]
for i, j in zip(a, b): # 집어 오는 함수
    print(f'a[{j}] = {i}')
```

```python
s = 'hello'
for idx, ch in enumerate(s): ## enumerate는 인덱스를 같이 줌, enumerate도 range처럼 클래스, 지금 생성자 호출 중
    print(f's[{idx}] = {ch}')
```

```python
list(enumerate(s))
```
