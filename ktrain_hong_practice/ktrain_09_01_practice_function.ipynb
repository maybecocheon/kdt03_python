```python
# STEP 1. 리스트 처리 함수 (기초 + comprehension)
# 과제 1
# 점수 리스트를 받아 70점 이상이면 '합격', 그렇지 않으면 '불합격'으로 분류된 문자열 리스트를 반환하세요.

# 입력 예시: [85, 67, 90, 45]
# 출력: ['합격', '불합격', '합격', '불합격']

def classify_scores(scores):
    return ['합격' if s >= 70 else '불합격' for s in scores]

print(classify_scores([85, 67, 90, 45]))
```

```python
# 과제 2
# 이름 리스트를 받아 각 이름에 순번을 붙인 리스트를 반환하세요 (enumerate() 사용).

# 출력 예시: ['1. 홍길동', '2. 김유신', ...]

def numbered_names(names):
    return [f'{i}. {name}' for i, name in enumerate(names, 1)]

print(numbered_names(['홍길동', '김유신', '강감찬', '계백']))
```

```python
# STEP 2. 튜플 & zip() 처리 함수
# 과제 3
# 이름, 점수 두 리스트를 받아 "홍길동: 85점"과 같은 문자열 리스트를 반환하세요 (zip() 사용).

def combine_name_score(names, scores):
    return [f'{name}: {score}점' for name, score in zip(names, scores)]

print(combine_name_score(['홍길동', '김유신', '강감찬', '계백'], [90, 75, 62, 10]))
```

```python
# 과제 4
# (이름, 점수) 튜플 리스트에서 80점 이상인 사람만 필터링된 튜플 리스트를 반환하세요.

def filter_high_scores(data):
    # return [d for d in data if d[1] >= 80]
    return list(filter(lambda d: d[1] >= 80, data))

print(filter_high_scores([('홍길동', 90), ('김유신', 75), ('강감찬', 62), ('계백', 10)]))
```

```python
# STEP 3. 딕셔너리 생성/변환 함수
# 과제 5
# 이름 리스트와 점수 리스트를 받아 딕셔너리로 변환하는 함수 (zip() 사용):

# 입력: ["홍길동", "김유신"], [85, 90]
# 출력: {"홍길동": 85, "김유신": 90}

def to_score_dict(names, scores):
    return dict(zip(names, scores))

print(to_score_dict(["홍길동", "김유신"], [85, 90]))
```

```python
# 과제 6
# 학생 점수 딕셔너리를 받아 합격자 이름만 리스트로 반환하세요 (cutoff 지정 가능).

def passed_students(score_dict, cutoff=70):
    return [key for key, value in score_dict.items() if value >= cutoff]

print(passed_students({"홍길동": 85, "김유신": 90, "계백": 65}))
```

```python
# STEP 4. 분석 보고용 함수
# 과제 7
# 이름, 과목, 점수 리스트를 받아 "1. 홍길동 - 수학: 85점"과 같은 리스트를 반환하세요 (enumerate(), zip() 함께 사용)

def formatted_report(names, subjects, scores):
    return [[f'{i}. {name} - {subject}: {score}점'] for i, (name, subject, score) in enumerate(zip(names, subjects, scores), 1)]

print(formatted_report(["홍길동", "김유신"], ["수학", "과학"], [90, 85]))
```

```python
# STEP 5. 집계 및 정제 함수
# 과제 8
# 문자열 리스트에서 공백 제거 + 소문자 변환 + 중복 제거된 정렬 리스트를 반환하세요.

def clean_unique_words(words):
    return list({s for s in words.replace(" ", "").lower()})

print(clean_unique_words("Abdes etsc eawgt"))
```

```python
# 과제 9
# 다수의 (이름, 점수) 튜플 리스트를 받아, 평균 점수와 최고 점수 학생을 출력하세요.

def score_summary(data):
    avgScore = sum(score for (_, score) in data) / len(data)
    maxScore = max(data, key=lambda x: x[1])
    return f'평균: {avgScore :.1f}, 최고 점수 학생: {maxScore}'

print(score_summary([("홍길동", 63), ("김유신", 75), ("계백", 80)]))
```

```python
# STEP 6: 확장 과제 (실전형)
# 과제 10: 데이터 전처리 → 분석까지 전 과정 함수 분리 실습
# 아래의 리스트를 기반으로 다음 작업을 각각의 함수로 분리하여 구현해보세요.

names = ["홍길동", " 김유신 ", "이순신", "강 감 찬"]
scores = [85, 70, 60, 90]
subjects = ["수학", "영어", "과학", "국어"]

# 함수 분리 목표
def clean_names(names): # → 이름에서 공백 제거 및 정리
    return [name.replace(" ", "") for name in names]

def to_dict(clean_names, scores): # → 이름과 점수 딕셔너리로 변환
    return {name:score for name, score in zip(clean_names, scores)}

def filter_pass(dic): # → 70점 이상인 학생만 필터링
    return list(filter(lambda x: x[1] >= 70, dic.items()))

def report(names, scores, subjects): # → "1. 홍길동 - 수학: 85점" 형태로 출력
    for i, (name, score, subject) in enumerate(zip(names, scores, subjects), 1):
        print(f'{i}. {name} - {subject}: {score}점')

name = clean_names(names)
print(name)
dic = to_dict(name, scores)
print(dic)
print(filter_pass(dic))
report(name, scores, subjects)
```
