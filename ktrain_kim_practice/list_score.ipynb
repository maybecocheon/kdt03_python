## 리스트를 사용한 성적 처리

```python
# 점수를 입력받아 총점과 평균 구하기 1
print('총점과 평균을 구해 봅시다.')
number = int(input('학생 수: '))

# score 리스트 초기화
score = [None] * number

# 각 학생의 점수 input
for i in range(number):
    score[i] = int(input(f'{i + 1}번 학생의 점수: '))
print(score)

# 학생들의 점수 더하기
result = 0
for i in score:
    result += i

# 총점과 평균 출력
print (f'총점은 {result}입니다.')
print (f'평균은 {result / number}입니다.')
```
