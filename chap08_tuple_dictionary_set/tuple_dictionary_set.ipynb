# 튜플, 딕셔너리, 집합
## 튜플

```python
# p. 322 튜플 생성
lst2 = list(range(5))
print('리스트:', lst2)

# 튜플 생성
tup = (1, 2, 3)
tup1 = tuple(range(5))
print('튜플:', tup1)
print('튜플 인덱스로 접근:', tup1[0])
```

```python
tup2 = (1,)     # 1, => 이렇게 써도 가능
print(type(tup2), tup2)
n = 1
print(type(n), n)
tup3 = 1, 2, 3  # pack 한다고 한다
print(type(tup3), tup3)
```

```python
# tuple() 함수 사용
tup1 = tuple([1, 2, 3])
tup2 = tuple({1, 2, 3})
tup3 = tuple(range(1, 4))

# divmod 함수 사용: 몫과 나머지를 튜플로 반환하는 함수
tup4 = divmod(13, 3)        
print(tup4)
d, r = divmod(13, 3)
tup5 = d, r
print(tup5)
```

```python
# p. 325 튜플 누적 대입
tp1 = (1, 2, 3)
print('tp1의 id값:', id(tp1))
tp1 += (4, 5)
print('누적 대입 결과 tp1:', tp1)
print('누적 대입 후 tp1의 id값:', id(tp1))

# 누적 대입하더라도 제자리 연산 불가
# 리스트는 누적 대입 시 제자리 연산
```

```python
# p. 327 튜플 정렬
# 튜플 정렬 => 원본 변경 안 함, 새로운 튜플 생성한 것
x = (1, 3, 2)
x = tuple(sorted(x))
print(x)
```

```python
# 튜플 대입
tpl1 = (1, 2, 3, 4, 5)
tpl2 = tuple(tpl1)
tpl3 = tpl1
print(id(tpl1))
print(id(tpl2))
print(id(tpl3))
```

```python
# p. 328 팩과 언팩
# 팩
tup4 = 1, 2, 3, 4 
print(tup4)

# 언팩
a, b, c, d = tup4
print('a:', a, 'b:', b, 'c:', c, 'd:', d)
e, f, _, _ = tup4
print('e:', e, 'f:', f, '[나머지는 언더바로 생략 => 익명 변수]')

# 팩 먼저 하고 다음 언팩 한다
x, y, z = 1, 2, 3
print(x, y, z)
```

```python
# 튜플이 있기 때문에 변수값 동시에 교환 가능
a, b = b, a
print('a:', a, 'b:', b)
```

```python
# p. 330 언팩의 리스트화
a, *b, c = tup4
print(a, b, c)

tup5 = 1, 2, 3, 4, 5
a, *b, c = tup5
print(a, b, c)
a, b, c, d, *e, f = tup5
print(a, b, c, d, e, f)
```

```python
# p. 331 enumerate 함수로 탐색하기 => enumerate는 튜플(이터러블) 객체 생성
s = 'abcde'
for i, ch in enumerate(s): # i, ch 변수에 생성된 튜플 객체를 언팩 한 것
    print(f's[{i}] = {ch}')
```

```python
# p. 332 튜플의 리스트
# 튜플도 인덱스로 접근 가능(인덱싱 가능)
students = [(2012, '김서준'), (2013, '이서아'), (2011, '최하준')]
print('students =', students)
print('students[0] =', students[0])
print('students[0][1] =', students[0][1])
```

```python
# p. 334 zip 함수의 결합
p1 = [1, 2, 3, 4]
p2 = [100, 200, 300]
name = ['홍길동', '강감찬', '김유신', '계백']
z = zip(p1, p2, name) # 인덱스 같은 것끼리 팩
print(z)
print(list(z))  # 리스트
pt = tuple(zip(p1, name, p2))
print(pt)       # 튜플
```

```python
num = range(6)
print(num)      # num은 해당 이터러블 객체를 가리키는 것에 불과하기 때문에 리스트나 튜플 등으로 만들어 주어야 값을 읽을 수 있음
print(list(num))
```

## 딕셔너리
- 키와 밸류로 이루어짐 (ex. JSON)

```python
# p. 338 딕셔너리는 키를 인덱스로 사용함
season = {
    '봄': 'spring',
    '여름': 'summer',
    '가을': 'autumn',
    '겨울': 'winter'
}
print('season["봄"] = ', season['봄'])
print(season['가을겨울'])       # 없는 키로 접근 시 에러 => 프로그램 중지
```

```python
# p. 342 딕셔너리 생성
# 1. 중괄호 사용(키: 밸류)
dic = {12: '이승엽', 18: '박민우'}      # 키는 불변 객체면 가능
print('중괄호 사용 생성:', dic)

# 2. dict 함수
lst = [('봄', 'spring'), ('여름', 'summer')]
dic2 = dict(lst)
print('dict 함수로 생성:', dic2)
# 2-1. dict 함수와 zip 함수
key = ['봄', '여름']
value = ['spring', 'summer']
print('dict 함수와 zip(리스트)로 생성:', dict(zip(key, value)))

key = ('봄', '여름')
value = ('spring', 'summer')
print('dict 함수와 zip(튜플)로 생성:', dict(zip(key, value)))
```

```python
# p. 343 등가성 판정
lst1 = ['korea', 'japan']
lst2 = ['japan', 'korea']
print('lst1 == lst2:', lst1 == lst2) # 리스트, 튜플, 문자열은 순서 다르면 등가 성립 x

dic1 = {'kor': 'korea', 'jap': 'japan'}
dic2 = {'jap': 'japan', 'kor': 'korea'}
print('dic1 == dic2:', dic1 == dic2) # 딕셔너리는 순서 달라도 요소(키와 값의 쌍)가 동일하다면 등가 성립 ㅇ => 순서 없기 때문에 인덱스 0, 1 등으로 접근 불가, 무조건 key로 접근

# 없는 키로 접근
print('없는 키로 접근 시 get 사용: dic1.get("usa") =', dic1.get('usa'))

# 딕셔너리 인덱스식
# 1. 없는 키면 값 추가
dic1['chn'] = 'china'
print(dic1['chn'])
# 2. 있는 키면 값 수정
dic1['kor'] = 'south korea'
print(dic1['kor'])
```

```python
# p. 347 setdefault 메서드
color = {'red': '빨강', 'green': '초록'}
print(color)
x = color.setdefault('red', '레드')     # 키 있으면 값 변경되지 않고 반환만 함
print(x)

y = color.setdefault('yellow', '노랑')  # 키 없으면 요소 추가함
print(y)

print(color)

# update 메서드
# 키가 있으면 수정, 키가 없으면 요소 추가
# 1. 딕셔너리
color.update({'blue': '파랑', 'red': '레드'})      # color 딕셔너리에 인수 딕셔너리 결합
print(color)

# 2. 키='값'
color.update(cyan = '하늘')         # 함수의 매개변수 전달 방식 중 '키워드 전달 방식'
print(color)

# 3. 반복 가능한 이터러블 객체
color.update((('magenta', '자주'), ('black', '검정')))
print(color)
color.update([('pink', '분홍')])
print(color)
```

```python
# p.351 요소 제거
color = {'red': '빨강', 'green': '초록'}
x = color.pop('red')
print(x)
```

```python
# p. 352 귀속 판정 연산자
if 'purple' not in color:
    color['purple'] = '보라'
print(color)
```

```python
# p. 353 딕셔너리 탐색
color = {'red': '빨강', 'green': '초록'}

# 1. enumerate 함수
for i, key in enumerate(color):         # 인덱스와 key가 나옴 => value는 안 나옴
    print(f'color[{i}] = {key}') 

# 2. for in
for key in color:
    print(key)
```

```python
# p. 354 딕셔너리와 뷰
color = {'red': '빨강', 'green': '초록'}
print('키:', tuple(color.keys()))       # 키의 뷰를 튜플로 변환
print('값:', tuple(color.values()))     # 값의 뷰를 튜플로 변환
print('요소:', tuple(color.items()))    # (키, 값)의 뷰를 튜플로 변환
```

```python
# p. 359 딕셔너리 내포 표기
txt = 'abcabcde'
count = { ch: txt.count(ch) for ch in txt }
print(count)

s = ['봄', '여름', '가을', '겨울']
season = { k: v for k, v in enumerate(s) }
print(season)

s = {0: '봄', 1: '여름', 2: '가을', 3: '겨울'}
season = { k: v for k, v in s.items() }
print(season)
```

## 집합

```python
# p.362 집합 생성
# 1. 중괄호
set1 = {}           # 빈 딕셔너리 생성
print(type(set1))
set2 = {1}
print(type(set2))

# 2. set 함수
set3 = set()
print(type(set3))
set3 = set([1, 2, 3, 2])    # set은 중복 불허
print(set3)
```

```python
# p. 363 집합은 가변 객체
set3 = set([1, 2, 3, 2])
print(id(set3))

# union 메서드: 합집합
# set3 = set3 + { 4, 5 } => 이렇게 합치면 오류
set3 = set3.union({ 4, 5 })
print(set3)
print(id(set3))

# add 메서드: 하나의 요소 집합에 추가
set4 = set([1, 2, 3, 2])           # 같은 값 가진 집합이면 참조변수 달라도 같은 집합 가리킴 (id값 동일)
print(id(set4))
set4.add(4)
print(set4)
print(id(set4))
```

```python
# 귀속 판정 연산자
set3 = set([1, 2, 3, 2])
n = 0
m = 2
if n in set3 and m in set3:
    print('사이클')
```
