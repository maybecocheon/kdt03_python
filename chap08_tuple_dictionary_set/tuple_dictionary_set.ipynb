# 튜플, 딕셔너리, 집합
## 튜플

```python
lst2 = list(range(5))
print('리스트:', lst2)

# 튜플 생성
tup = (1, 2, 3)
tup1 = tuple(range(5))
print('튜플:', tup1)
print('튜플 인덱스로 접근:', tup1[0])
```

```python
tup2 = (1,)     # 1, => 이렇게 써도 가능
print(type(tup2), tup2)
n = 1
print(type(n), n)
tup3 = 1, 2, 3  # pack 한다고 한다
print(type(tup3), tup3)
```

```python
# tuple() 함수 사용
tup1 = tuple([1, 2, 3])
tup2 = tuple({1, 2, 3})
tup3 = tuple(range(1, 4))

# divmod 함수 사용: 몫과 나머지를 튜플로 반환하는 함수
tup4 = divmod(13, 3)        
print(tup4)
d, r = divmod(13, 3)
tup5 = d, r
print(tup5)
```

```python
tp1 = (1, 2, 3)
print('tp1의 id값:', id(tp1))
tp1 += (4, 5)
print('누적 대입 결과 tp1:', tp1)
print('누적 대입 후 tp1의 id값:', id(tp1))

# 누적 대입하더라도 제자리 연산 불가
# 리스트는 누적 대입 시 제자리 연산
```

```python
# 튜플 정렬 => 원본 변경 안 함, 새로운 튜플 생성한 것
x = (1, 3, 2)
x = tuple(sorted(x))
print(x)
```

```python
# 튜플 대입
tpl1 = (1, 2, 3, 4, 5)
tpl2 = tuple(tpl1)
tpl3 = tpl1
print(id(tpl1))
print(id(tpl2))
print(id(tpl3))
```

```python
# 튜플 팩
tup4 = 1, 2, 3, 4 
print(tup4)

# 언팩
a, b, c, d = tup4
print('a:', a, 'b:', b, 'c:', c, 'd:', d)
e, f, _, _ = tup4
print('e:', e, 'f:', f, '[나머지는 언더바로 생략 => 익명 변수]')

# 팩 먼저 하고 다음 언팩 한다
x, y, z = 1, 2, 3
print(x, y, z)
```

```python
# 튜플이 있기 때문에 변수값 동시에 교환 가능
a, b = b, a
print('a:', a, 'b:', b)
```
